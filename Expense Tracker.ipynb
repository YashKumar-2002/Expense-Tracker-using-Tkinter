```python
import tkinter as tk
from tkinter import ttk, messagebox

total_received = 0
total_spent = 0
history = []

def add_received():
    try:
        global total_received
        amount_received = float(received_entry.get())
        source_received = received_source_entry.get()
        total_received += amount_received
        history.append((amount_received, source_received, "Received"))
        received_entry.delete(0, 'end')  # Clear the entry field
        received_source_entry.delete(0, 'end')  # Clear the entry field
        update_display()
    except ValueError:
        messagebox.showerror("Error", "Invalid input. Please enter a valid amount.")

def add_spent():
    try:
        global total_spent
        amount_spent = float(spent_entry.get())
        source_spent = spent_source_entry.get()
        total_spent += amount_spent
        history.append((amount_spent, source_spent, "Spent"))
        spent_entry.delete(0, 'end')  # Clear the entry field
        spent_source_entry.delete(0, 'end')  # Clear the entry field
        update_display()
    except ValueError:
        messagebox.showerror("Error", "Invalid input. Please enter a valid amount.")

def update_display():
    total_received_label.config(text=f"Total Received: ${total_received:.2f}")
    total_spent_label.config(text=f"Total Spent: ${total_spent:.2f}")
    remaining_balance = total_received - total_spent
    balance_label.config(text=f"Remaining Balance: ${remaining_balance:.2f}")

def show_history():
    history_window = tk.Toplevel(root)
    history_window.title("Expense History")

    ttk.Label(history_window, text="History").grid(row=0, column=0, columnspan=3, pady=10)

    ttk.Label(history_window, text="Amount").grid(row=1, column=0, padx=5, pady=5)
    ttk.Label(history_window, text="Source").grid(row=1, column=1, padx=5, pady=5)
    ttk.Label(history_window, text="Type").grid(row=1, column=2, padx=5, pady=5)

    row_num = 2
    for entry in history:
        ttk.Label(history_window, text=f"${entry[0]:.2f}").grid(row=row_num, column=0, padx=5, pady=5)
        ttk.Label(history_window, text=entry[1]).grid(row=row_num, column=1, padx=5, pady=5)
        ttk.Label(history_window, text=entry[2]).grid(row=row_num, column=2, padx=5, pady=5)
        row_num += 1

root = tk.Tk()
root.title("Expense Tracker")

# Entry fields
received_label = ttk.Label(root, text="Amount Received:")
received_label.grid(row=0, column=0, padx=10, pady=5, sticky="w")
received_entry = ttk.Entry(root)
received_entry.grid(row=0, column=1, padx=10, pady=5, sticky="w")

received_source_label = ttk.Label(root, text="Source:")
received_source_label.grid(row=0, column=2, padx=10, pady=5, sticky="w")
received_source_entry = ttk.Entry(root)
received_source_entry.grid(row=0, column=3, padx=10, pady=5, sticky="w")

spent_label = ttk.Label(root, text="Amount Spent:")
spent_label.grid(row=1, column=0, padx=10, pady=5, sticky="w")
spent_entry = ttk.Entry(root)
spent_entry.grid(row=1, column=1, padx=10, pady=5, sticky="w")

spent_source_label = ttk.Label(root, text="Source:")
spent_source_label.grid(row=1, column=2, padx=10, pady=5, sticky="w")
spent_source_entry = ttk.Entry(root)
spent_source_entry.grid(row=1, column=3, padx=10, pady=5, sticky="w")

# Buttons
add_received_button = ttk.Button(root, text="Add Received", command=add_received)
add_received_button.grid(row=2, column=0, padx=10, pady=5, sticky="w")

add_spent_button = ttk.Button(root, text="Add Spent", command=add_spent)
add_spent_button.grid(row=2, column=1, padx=10, pady=5, sticky="w")

show_history_button = ttk.Button(root, text="Show History", command=show_history)
show_history_button.grid(row=2, column=2, padx=10, pady=5, sticky="w")

# Display labels
total_received_label = ttk.Label(root, text="Total Received: ₹0.00")
total_received_label.grid(row=3, column=0, padx=10, pady=5, sticky="w")

total_spent_label = ttk.Label(root, text="Total Spent: ₹0.00")
total_spent_label.grid(row=4, column=0, padx=10, pady=5, sticky="w")

balance_label = ttk.Label(root, text="Remaining Balance: ₹0.00")
balance_label.grid(row=5, column=0, padx=10, pady=5, sticky="w")

root.mainloop()
```
